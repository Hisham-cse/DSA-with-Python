```python
## Largest Element in a list
def largest(l):
    for i in l:
        for j in l:
            if j>i:
                break
        else:
            return i
    return None

lst=[10,20,30,40,50]
print(largest(lst))

# Time O(n**2)
```

```python
def getMax(l):
    if not l:
        return None
    else:
        res=l[0]
        for i in range(1,len(l)):
             if l[i]>res:
                 res = l[i]
        return res
l=[2,1,3,4,0]
print(getMax(l))

# Time O(n)
```
